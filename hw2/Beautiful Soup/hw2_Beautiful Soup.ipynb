```python
import requests
from bs4 import BeautifulSoup
import csv

# 爬取的網頁
url = "https://school.chc.edu.tw/user/basic/view?all=true"
response = requests.get(url)
html_content = response.content

# 使用 BeautifulSoup 解析 HTML 内容
soup = BeautifulSoup(html_content, "html.parser")

# 找到要爬取的表格
table = soup.find("table")

# 創一個csv檔
csv_filename = "output_bs.csv"
with open(csv_filename, "w", newline="", encoding="utf-8-sig") as csv_file:
    csv_writer = csv.writer(csv_file)

    # 表格標題列
    header = [th.text for th in table.find_all("th")]
    csv_writer.writerow(header)

    # 表格內容
    for row in table.find_all("tr"):
        csv_writer.writerow([td.text for td in row.find_all("td")])
```
