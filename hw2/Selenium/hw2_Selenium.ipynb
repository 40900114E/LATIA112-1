```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

# 設定 ChromeOptions
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 建立 Chrome WebDriver 並傳入 options 參數
driver = webdriver.Chrome(options=options)

url = "https://school.chc.edu.tw/user/basic/view"
driver.get(url)

table = driver.find_element(By.TAG_NAME, 'table')

# 算出總共有幾頁
pagination = driver.find_element(By.XPATH, '//*[@id="schoolTableUl"]')
pages = pagination.find_elements(By.TAG_NAME, 'a')
last_page = int(pages[-2].text)
print("totalPage", last_page)

# 建立一個陣列存放要放進最後檔案的資料
table1 = [["#", "學校名稱", "地址", "電話", "網路電話", "傳真", "學校首頁", "更新時間"]]

for currentpage in range(1, last_page):
    for i, row in enumerate(driver.find_elements(By.TAG_NAME, "tr")):
        rowData = []
        for column in row.find_elements(By.TAG_NAME, "td"):
          rowData.append(column.text)
        # 檢查該行是否全部為空白
        if any(rowData):
          table1.append(rowData)

    next_page = driver.find_element(By.XPATH, '//*[@id="b_next"]')
    next_page.click()
    print("currentpage", currentpage)
    # 避免伺服器壓力過大
    sleep(1)

sleep(5)
driver.close()

print(table1)
```

```python
import csv

csv_filename = "output_se.csv"
with open(csv_filename, "w", newline="", encoding="utf-8-sig") as csv_file:
  writer = csv.writer(csv_file)

  # 寫進表格
  writer.writerows(table1)
```
