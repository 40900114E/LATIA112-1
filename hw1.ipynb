```python
# 套件載入
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# plt顯示中文設定
# 下載台北思源黑體
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')
```

```python
# 讀取資料
df = pd.read_csv('111_basec.csv', encoding='UTF-8')
# 列印資料前5行
df.head(n=5)
```

### **第一部分：學校概況**

Q1. 國民小學在各縣市的數量？哪個縣市的國民小學最多？哪個縣市的國民小學最少？

```python
# 計算各縣市的出現個數
city_list = df["縣市名稱"].unique()
city_count = df["縣市名稱"].value_counts()

print(f"111學年臺灣共有{len(df['學校代碼'])}所國民小學，其中：\n")
for i in city_list:
  print(f"{i}有{city_count[i]}所")

# 找出最大值
city_maxname = city_count.idxmax()
city_max = city_count.max()

# 找出最小值
city_minname = city_count.idxmin()
city_min = city_count.min()
print(f"\n國民小學最多的縣市是{city_maxname}，有{city_max}所;\n國民小學最少的縣市是{city_minname},有{city_min}所。\n")

# 繪製長條圖
sizes = [city_count[x] for x in city_list]
plt.barh(city_list,sizes,color = mcolors.TABLEAU_COLORS)
plt.title("各縣市國民小學數量長條圖")
plt.xlabel("國民小學數量（所）")
plt.ylabel("縣市")
for y in range(len(city_list)):
  plt.text(sizes[y]+3,y,sizes[y],va='center',fontsize=8)
plt.show()
```

Q2. 六個直轄市中的國民小學合計有多少所？非六個直轄市中的國民小學合計又有多少所？

```python
# 將縣市名稱替換為六都與非六都
spcity = df["縣市名稱"].replace({"新北市","臺北市","桃園市","臺中市","臺南市","高雄市"},"六都")
spcity_list = spcity.replace({"連江縣","金門縣","嘉義市","新竹市","基隆市","澎湖縣","花蓮縣","臺東縣","屏東縣","嘉義縣","雲林縣","南投縣","彰化縣","苗栗縣","新竹縣","宜蘭縣"},"非六都")
spcity_count = spcity_list.value_counts()

print(f"六個直轄市的國民小學合計有{spcity_count['六都']}所；\n非六個直轄市中的國民小學合計有{spcity_count['非六都']}所。\n")

# 繪製圓餅圖
sizes = [spcity_count["六都"],spcity_count["非六都"]]
labels = ["六都","非六都"]
plt.pie(sizes, labels = labels, autopct="%.1f%%")
plt.title("直轄市與非直轄市國民小學比例圖")
plt.show()
```

Q3. 111學年國民小學平均每校有多少班？班級數最多的是哪間學校？最少的又是哪間？

```python
# 算出各學校班級總數
class_list = df[[f"{i}年級班級數" for i in range(1,7)]].sum(1)

# 算出班級數的平均
class_mean = np.mean(class_list)

# 找出最大值
class_max = class_list.max()
class_maxname = df["學校名稱"][class_list.idxmax()]

# 找出最小值
class_min = class_list.min()
class_minname = df["學校名稱"][class_list.idxmin()]

print(f"111學年國民小學平均每校有{round(class_mean)}班（四捨五入至整數位），其中：\n\n班級數最多的是{class_maxname}，有{class_max}班；\n班級數最少的是{class_minname}，有{class_min}班。")
```

Q4. 111學年國民小學中屬於「附設國小」型態的有幾所？佔全部國民小學多少比例？

```python
# 找出所有學校名稱內含有「附設國小」的資料
schoolsp_count = df[df["學校名稱"].str.contains("附設國小")]
# 算出比例
schoolsp_proportion = round((len(schoolsp_count)/(len(df["學校代碼"])-len(schoolsp_count)))*100,2)

print(f"111學年國民小學中屬於「附設國小」型態的有{len(schoolsp_count)}所，佔全部的{schoolsp_proportion}%（四捨五入至小數第二位）；\n不屬於「附設國小」型態的有{len(df['學校代碼'])-len(schoolsp_count)}所。\n\n")

# 繪製圓餅圖
sizes = [len(schoolsp_count),len(df['學校代碼'])-len(schoolsp_count)]
labels = ["附設國小","非附設國小"]
plt.pie(sizes, labels = labels, autopct="%.1f%%")
plt.title("附設國小與非附設國小比例圖")
plt.show()
```

### **第二部分：學生概況**

Q5. 111學年國民小學學生總數為多少人？各年級又有多少人？

```python
# 將各年級男學生數與女學生數相加為各年級學生數
student_grade = [df[f"{i}年級男學生數"].sum()+df[f"{i}年級女學生數"].sum() for i in range (1,7)]

# 將各年級學生數加總，得到學生總數
student_sum = sum(student_grade)
print(f"111學年國民小學學生總數為{student_sum}人，其中：\n")
for x in range(1,7):
  print(f"{x}年級學生共有{student_grade[x-1]}人")

# 繪製長條圖
labels = [f"{y}年級" for y in range(1,7)]
plt.bar(labels, student_grade, color = mcolors.TABLEAU_COLORS)
plt.title("111學年國民小學各年級學生數長條圖")
plt.xlabel("年級")
plt.ylabel("學生人數（人）")
for a in range(len(labels)):
  plt.text(a,student_grade[a]-10000,student_grade[a],ha='center',fontsize=9, color="w")
plt.show()
```

Q6. 111學年國民小學男學生總數為多少人？女學生又有多少人？性別比為多少？

```python
# 將各年級男學生數相加
student_male = sum([df[f"{i}年級男學生數"].sum() for i in range (1,7)])

# 將各年級女學生數相加
student_female = sum([df[f"{i}年級女學生數"].sum() for i in range (1,7)])

# 性別比
sex_ratio = round(student_male/student_female*100)

print(f"111學年國民小學男學生總數為{student_male}人；\n111學年國民小學女學生總數為{student_female}人。\n\n性別比為{sex_ratio}（四捨五入至整數位）。\n\n")

# 繪製圓餅圖
sizes = [student_male,student_female]
labels = ["男學生數","女學生數"]
plt.pie(sizes, labels = labels, autopct="%.1f%%")
plt.title("111學年國民小學學生性別比例圖")
plt.show()
```

Q7. 111學年國民小學中學生數量最多的是哪所學校？最少的是哪所學校？

```python
# 算出各學校的學生總數
student_school = df[[f"{i}年級男學生數" for i in range(1,7)]].sum(1)+df[[f"{i}年級女學生數" for i in range(1,7)]].sum(1)

# 最大值
student_max = student_school.max()
student_maxname = df["學校名稱"][student_school.idxmax()]

# 最小值
student_min = student_school.min()
student_minname = df["學校名稱"][student_school.idxmin()]

print(f"111學年國民小學中學生數量最多的是{student_maxname}，有{student_max}位學生；\n學生數量最少的是{student_minname}，有{student_min}位學生。")
```

### **第三部分：教師概況**

Q8. 111學年國民小學專任教師總數為多少人？其中男專任教師有多少人？女專任教師有多少人？

```python
# 分別計算男、女專任教師人數
teacher_male = sum(df["男專任教師"])
teacher_female = sum(df["女專任教師"])

# 加總成為專任教師總數
teacher_count = teacher_male+teacher_female

print(f"111學年國民小學專任教師總數為{teacher_count}人，其中：\n男專任教師有{teacher_male}人，女專任教師有{teacher_female}人。\n\n")

# 繪製圓餅圖
sizes = [teacher_male,teacher_female]
labels = ["男專任教師數","女專任教師數"]
plt.pie(sizes, labels = labels, autopct="%.1f%%")
plt.title("111學年國民小學專任教師性別比例圖")
plt.show()
```

Q9. 111學年國民小學平均每所學校有幾位專任教師？專任教師最多的是哪所學校？最少的又是哪所學校？

```python
# 每間學校的專任教師加總
teacher_school = df[["男專任教師","女專任教師"]].sum(1)
teacher_mean = round(np.mean(teacher_school))

# 最大值
teacher_max = teacher_school.max()
teacher_maxname = df["學校名稱"][teacher_school.idxmax()]

# 最小值
teacher_min = teacher_school.min()
teacher_minname = df["學校名稱"][teacher_school.idxmin()]

print(f"111學年國民小學平均每所學校有{teacher_mean}位專任教師（四捨五入至整數位），其中：\n\n專任教師最多的學校是{teacher_maxname}，有{teacher_max}位；\n專任教師最少的學校是{teacher_minname}，有{teacher_min}位。")
```

Q10. 111學年國民小學平均的生師比為多少？生師比最大的是哪所學校？

```python
# 用學生數量除以專任教師數量，得到每間學校的生師比
student_teacher_ratio = round(student_school/teacher_school)
# 因資料教師人數裡有0，相除後會是inf，把inf的資料變成NAN
where_are_inf = np.isinf(student_teacher_ratio)
student_teacher_ratio[where_are_inf]=0

# 平均
student_teacher_ratio_mean = round(np.mean(student_teacher_ratio))

# 最大值
student_teacher_ratio_max = student_teacher_ratio.max()
student_teacher_ratio_maxname = df["學校名稱"][student_teacher_ratio.idxmax()]

print(f"111學年國民小學平均的生師比為{student_teacher_ratio_mean}（四捨五入至整數位），代表每1位專任教師平均約要教導{student_teacher_ratio_mean}位學生；\n生師比最大的學校是{student_teacher_ratio_maxname}，生師比為{student_teacher_ratio_max}，代表每1位專任教師平均約要教導{student_teacher_ratio_max}位學生。")

# 繪製長條圖
sizes = [student_teacher_ratio_mean, student_teacher_ratio_max]
labels = ["全國平均生師比",student_teacher_ratio_maxname+"生師比"]
plt.bar(labels, sizes, color = mcolors.TABLEAU_COLORS)
plt.title("111學年國民小學生師比長條圖")
plt.xlabel("比較場域")
plt.ylabel("生師比")
for a in range(len(labels)):
  plt.text(a,sizes[a]-2,sizes[a],ha='center',fontsize=14, color="w")
plt.show()
```
