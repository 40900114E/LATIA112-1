```python
# 套件載入
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# plt顯示中文設定
plt.rc('font', family='Microsoft JhengHei')
```

```python
# 讀取資料
df=pd.read_csv('111_students.csv', encoding='UTF-8')
# 列印資料前5行
df.head(n=5)
```

### **第一部分：學校概況**

Q1.大專院校分為哪些體系？各體系學校數量為何？

```python
from pandas.core.algorithms import value_counts
# 取出資料中不重複的體系別
type_list = list(np.unique(df["體系別"]))
# 按照編號排序
type_list = sorted(type_list)
print("大專院校體系分為：")
for i in type_list:
  print(f"「{i}」",end = "")

# 計算各體系類別的數量，並按照編號排序
type_count = df["體系別"].value_counts().sort_index(ascending=True)
print(f"\n\n各體系學校數量：")
for i in type_list:
  print(f"「{i}」有{type_count[i]}所")

plt.pie(type_count,labels=type_list,explode=type_count)
plt.title("111學年大專院校各體系比例圖")
plt.show()
```
